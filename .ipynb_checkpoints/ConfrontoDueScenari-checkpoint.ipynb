```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
# carica dati
#data = pd.read_csv('/Users/alket/Desktop/dati/new_data_Modena_backfill_forwfill.csv', index_col = 0, header=0, parse_dates=True)
data = pd.read_csv('/Users/alket/Desktop/dati/new_data_backfill_forwfill.csv', index_col = 0, 
                   header=0, parse_dates=True)

# aggrega dati
agg_by_cell = data.groupby(by = ['cell_num'])
```

```python
counter = 0
for index, frame in agg_by_cell: 
    print(index)
    counter += 1
    if counter > 3: break
    frame2 = frame.iloc[::4,:]
    print(len(frame2))
    plt.figure(figsize = (20,4))
    plt.plot(frame2['nr_people'])
    plt.show()
```

```python
cells2var = {}
counter = 0
cells = []
mean_cell = []
std_cell = []
for index, frame in agg_by_cell:
    #print(index, frame)
    counter += 1
    if counter > 2: break
    frame = frame[44:]
    cell_i = frame.iloc[::4,:]['nr_people'].values
    print(frame.iloc[::4,:].head(5))
    mean = np.mean(cell_i)
    std = np.std(cell_i)
    
    cells.append(index)
    mean_cell.append(mean)
    std_cell.append(std)
    percent_change = 100 * (frame.iloc[::94,:]['nr_people'].pct_change())
    #print(type(percent_change))  
    std_dev = percent_change.std()
    #print(std_dev)
    cells2var[index] = std_dev
    #print('cell mean and std', index, mean, std)
```

```python
cells2var
```

```python
std_dist = list(cells2var.values())
plt.figure(figsize = (20, 4))
plt.hist(std_dist, bins = 40)
plt.show()
```

```python
cells2con = []
for key, val in cells2var.items():
    if val > 15.5: 
        cells2con.append(key)
print(len(cells2con))
print(cells2con)
```

```python
percent_std_cell = np.array(std_cell) / np.array(mean_cell)
per_std_cell = percent_std_cell.tolist()
```

```python
norm = np.linalg.norm(np.array(percent_std_cell) )
per_normal = np.array(std_cell)/norm
```

```python
data_dist = pd.DataFrame({'cell':cells, 'mean':mean_cell, 'stddev':std_cell, 'per_mean_norm':per_normal})
data_dist.head(50)
```

```python
plt.figure(figsize = (20, 4))
plt.hist(data_dist['per_mean_norm'], bins = 100)
plt.show()
```

```python
anormal = data_dist[data_dist['per_mean_norm']>= 10]
len(anormal)
```
