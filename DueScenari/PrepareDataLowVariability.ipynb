```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/alket/Desktop/dati/new_data_backfill_forwfill.csv', index_col = 0, 
                   header=0, parse_dates=True)
```

```python
# aggrega dati
agg_by_cell = data.groupby(by = ['cell_num'])
```

```python
cells2var = {}
counter = 0
cells = []
mean_cell = []
std_cell = []
for index, frame in agg_by_cell:
    #print(index, frame)
    counter += 1
    #if counter > 2: break
    frame = frame[44:]
    cell_i = frame.iloc[::4,:]['nr_people'].values
    print(frame.iloc[::4,:].head(5))
    mean = np.mean(cell_i)
    std = np.std(cell_i)
    
    cells.append(index)
    mean_cell.append(mean)
    std_cell.append(std)
    percent_change = 100 * (frame.iloc[::94,:]['nr_people'].pct_change())
    #print(type(percent_change))  
    std_dev = percent_change.std()
    #print(std_dev)
    cells2var[index] = std_dev
```

```python
std_dist = list(cells2var.values())
plt.figure(figsize = (20, 4))
plt.hist(std_dist, bins = 40)
plt.show()
```

```python
cells = []
for key, val in cells2var.items():
    if val < 9: 
        cells.append(key)
print(len(cells))
print(cells)
```
